#***after the basic analysis we have to remove some unnecessary columns*** 

```python
new_dataset = dataset[['id','name','host_id','host_name','neighbourhood_group','neighbourhood','room_type','price','minimum_nights','last_review',
             'number_of_reviews','calculated_host_listings_count','availability_365']]
new_dataset.head(5)
```

### ***Area mostly chosen by the host or visitor***

```python
hosts_areas = new_dataset.groupby(['host_name','neighbourhood_group'])['calculated_host_listings_count'].max().reset_index()
hosts_areas.sort_values(by='calculated_host_listings_count', ascending=False).head(5)
```

#***now we look in dataset  for another listed data i.e price, review,  for the analysis***

```python
areas_reviews = new_dataset.groupby(['neighbourhood_group'])['number_of_reviews'].max().reset_index()
areas_reviews
```

#***to understand the comparison of data we use visualization technique***

```python
area = areas_reviews['neighbourhood_group']
review = areas_reviews['number_of_reviews']

fig = plt.figure(figsize = (12, 8))
 
# creating the bar plot
plt.bar(area, review, color ='Red',
        width = 0.4)
 
plt.xlabel("area")
plt.ylabel("review")
plt.title("Area vs Number of reviews")
plt.show()
```

#***now take a look relation between the number of review and price***

```python
price_area = new_dataset.groupby(['price'])['number_of_reviews'].max().reset_index()
price_area.head(5)
```

```python
area = price_area['price']
price = price_area['number_of_reviews']

fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.scatter(area, price)
 
plt.xlabel("Price")
plt.ylabel("Number of Review")
plt.title("Price vs Number of Reviews")
plt.show()
```

#***from the above plot we can clearly see that people stay in hotel with highest rating with cheap price.***


#Which hosts are the busiest and why?

```python
busiest_hosts = new_dataset.groupby(['host_name','host_id','room_type'])['number_of_reviews'].max().reset_index()
busiest_hosts = busiest_hosts.sort_values(by='number_of_reviews', ascending=False).head(10)
busiest_hosts
```

```python
name = busiest_hosts['host_name']
reviews = busiest_hosts['number_of_reviews']

fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(name, reviews, color ='red',
        width = 0.4)
 
plt.xlabel("Name of the Host")
plt.ylabel("Number of Reviews")
plt.title("Busiest Hosts")
plt.show()
```

#***most choosen room type and the reason***

```python
traffic_areas = new_dataset.groupby(['neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
traffic_areas = traffic_areas.sort_values(by='minimum_nights', ascending=False)
traffic_areas
```

```python
room_type = traffic_areas['room_type']
stayed = traffic_areas['minimum_nights']

fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(room_type, stayed, color ='red',
        width = 0.4)
 
plt.xlabel("Room Type")
plt.ylabel("Minimum number of nights stayed")
plt.title("Traffic Areas")
plt.show()
```

#*From the Above Analysis We can Stay that People are preferring Entire home/apt or Private room which are present in Manhattan, Brooklyn, Queens and people are preferring listings which are less in price.*
